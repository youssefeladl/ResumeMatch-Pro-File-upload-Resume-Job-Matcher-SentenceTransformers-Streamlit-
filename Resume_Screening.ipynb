#DATA PREPROCESSING

```python
import re
def clean_text(text):
  text=str(text).lower()
  text=re.sub(r'[^a-zA-Z ]', ' ', text)
  text=re.sub(r'\s+', ' ', text)
  return text.strip()

df_resume['cleaned_resume']=df_resume['Resume'].apply(clean_text)
df_job['cleaned_job_description']=df_job['Job Description'].apply(clean_text)
```

#MOODELING

```python
from sentence_transformers import SentenceTransformer
model=SentenceTransformer('sentence-transformers/all-mpnet-base-v2')
resume_text=df_resume['cleaned_resume'].tolist()
job_text=df_job['cleaned_job_description'].tolist()
resume_embedding=model.encode(resume_text)
job_embedding=model.encode(job_text)
model.save('my_sentence_model')
print('resume embedding',resume_embedding[0][0:10])
print('job embedding',job_embedding[0][0:10])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix=cosine_similarity(resume_embedding, job_embedding)
print('shape:',similarity_matrix.shape)
print(similarity_matrix[:10][:10])
```

```python
import pandas as pd
import numpy as np

top_k = 5
results = []

for resume_idx in range(similarity_matrix.shape[0]):
    sim_scores = similarity_matrix[resume_idx]

    top_jobs_indices = np.argsort(sim_scores)[::-1][:top_k].tolist()
    top_jobs_scores = sim_scores[top_jobs_indices].tolist()
    results.append({
        'resume_index': resume_idx,
        'top_jobs_indices': top_jobs_indices,
        'top_jobs_scores': top_jobs_scores
    })
df_top_jobs = pd.DataFrame(results)
print(df_top_jobs)
```

```python
all_skills=list(set(sum(job_skills.values(),[])))
def extract_skills(cleaned_resume,all_skills):
  found=[]
  for skill in all_skills:
    if skill in cleaned_resume:
      found.append(skill)
  return found
```

```python
def certification_extract(cleaned_resume,cert_keywords):
  found=[]
  for cert in cert_keywords:
    if cert in cleaned_resume:
      found.append(cert)
  return found
```

```python
def projects_extract(cleaned_resume,project_keywords):
  found=[]
  for project in project_keywords:
    if project in cleaned_resume:
      found.append(project)
  return found
```

```python
all_skills_found = sum(df_resume['skills_found'], [])
all_cert_found = sum(df_resume['certification_found'], [])
all_projects_found = sum(df_resume['projects_found'], [])
```

```python
all_skills_found = list(set(all_skills_found))
all_cert_found = list(set(all_cert_found))
all_projects_found = list(set(all_projects_found))
```

```python
df_resume['all_keywords'] = df_resume.apply(
    lambda row: row['skills_found'] + row['certification_found'] + row['projects_found'], axis=1
)
```

```python
df_resume['all_keywords_str'] = df_resume['all_keywords'].apply(lambda x: ' '.join(x))
```

```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
model = SentenceTransformer('all-MiniLM-L6-v2')
resume_embeddings = model.encode(df_resume['all_keywords_str'].tolist(), convert_to_tensor=False)
job_embeddings = model.encode(df_job['cleaned_job_description'].tolist(), convert_to_tensor=False)
model.save('my_sentence_model')
```

```python
cosine_scores = cosine_similarity(resume_embeddings, job_embeddings)
```

```python
def count_matches(resume_keywords, job_keywords):
    return len(set(resume_keywords) & set(job_keywords))
```

```python
skill_scores = []

for resume_skills in df_resume['skills_found']:
    row_skill_scores = []
    for job_keywords in df_job['cleaned_job_description']:
        row_skill_scores.append(count_matches(resume_skills, job_keywords))
    skill_scores.append(row_skill_scores)

skill_scores_df = pd.DataFrame(skill_scores)
```

```python
cert_scores = []
for resume_certs in df_resume['certification_found']:
    row_cert_scores = []
    for job_keywords in df_job['cleaned_job_description']:
         row_cert_scores.append(count_matches(resume_certs, job_keywords)) # Assuming count_matches works with lists of strings
    cert_scores.append(row_cert_scores)

cert_scores_df = pd.DataFrame(cert_scores)
```

```python
project_scores = []
for resume_projects in df_resume['projects_found']:
    row_project_scores = []
    for job_keywords in df_job['cleaned_job_description']:
         row_project_scores.append(count_matches(resume_projects, job_keywords))
    project_scores.append(row_project_scores)

project_scores_df = pd.DataFrame(project_scores)
```

```python
final_scores_df = (
    0.7 * pd.DataFrame(cosine_scores) +
    0.2 * skill_scores_df +
    0.1 * cert_scores_df +
    0.1 * project_scores_df
)
```

```python
top_k = 5
top_jobs_indices = final_scores_df.values.argsort(axis=1)[:, ::-1][:, :top_k]
top_jobs_scores = np.take_along_axis(final_scores_df.values, top_jobs_indices, axis=1)
```

```python
results = []
for i in range(len(df_resume)):
    results.append({
        'resume_index': i,
        'top_jobs_indices': top_jobs_indices[i].tolist(),
        'top_jobs_scores': top_jobs_scores[i].tolist()
    })

final_df = pd.DataFrame(results)
print(final_df.head())
```
